```python
x = np.array([3, 5])
y = np.array([2, 5])
print(np.greater_equal(x,y))
print(np.less_equal(x,y))
print(np.less(x,y))
```

```python
import numpy as np
x = np.array([72, 79, 85, 90, 150, -135, 120, -10, 60, 100])
y = np.array([72, 79, 85, 90, 150, -135, 120, -10, 60, 100.000001])
print(np.equal(x,y))
print(np.allclose(x,y))
```
