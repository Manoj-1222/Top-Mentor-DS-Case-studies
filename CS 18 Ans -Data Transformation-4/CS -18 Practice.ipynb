```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

from IPython.core.display import display_webp

plt.rcParams['figure.figsize'] = (20,10)
```

```python
data = pd.read_csv('BigMartSalesData.csv')
```

```python
data.size
```

```python
plt.plot(data['Month'],data['Amount'])
plt.show()
```

```python
monthly_sales = data.groupby('Month')['Amount'].sum()
print(monthly_sales)
```

```python
months = sorted(data['Month'].unique())
```

```python
plt.plot(months,monthly_sales,linewidth=5)
plt.xlabel("Month NUmber")
plt.ylabel("Sales in Millions of Euros")
for i in range(len(months)):
    plt.text(months[i],months[i],months[i])
plt.legend()
plt.show()
```

```python
# 2.Plot Total Sales Per Month for Year 2011 as Bar Chart.
# Is Bar Chart Better to visualize than Simple Plot?
plt.bar(months, monthly_sales)
plt.xticks(months)
for i in range(len(months)):
    plt.text(months[i], monthly_sales.iloc[i], f'{monthly_sales.iloc[i]:.2f}', ha='center', va='bottom')
plt.show()
```

#### 3.Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?

```python
country_wise_sales = data.groupby('Country')['Amount'].sum()
plt.pie(country_wise_sales,labels=country_wise_sales.index,autopct='%1.1f%%')
plt.title("Country wise Sales Contribution",loc='left')
plt.show()
```
