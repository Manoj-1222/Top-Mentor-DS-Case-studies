```python
X = [1,2,3,4]
Y= [20, 21, 20.5, 20.8]
plt.plot(X,Y,c='r',linewidth=2,linestyle='solid',marker='>',markersize=12,markerfacecolor='g')
plt.xlim(0.5,4.5)
plt.xticks([1,2,3,4])
plt.ylim(19.5,22)
plt.yticks([20,21,22,23,24])
plt.title("Graph 1")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
fig = plt.figure(figsize=(4,5),dpi=100)
x = [1,2,3,4]
y= [20,21,20.5,20.8]
y_error = [0.12,0.13,0.2,0.1]
plt.plot(x,y,linestyle='dashed',c='green',marker='o')
plt.rc("font",size=14)
plt.errorbar(x,y,yerr=y_error,linestyle='solid',marker='None',ecolor='red',color='g')
```

```python
# 4.9: Create a dataframe from following data
#
#
# 'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
#
# 'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
#
# 'female': [0, 1, 1, 0, 1],
#
# 'age': [42, 52, 36, 24, 73],
#
# 'preTestScore': [4, 24, 31, 2, 3],
#
# 'postTestScore': [25, 94, 57, 62, 70]
#
#
# Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
```

```python
df = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],

'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],

'female': [0, 1, 1, 0, 1],

'age': [42, 52, 36, 24, 73],

'preTestScore': [4, 24, 31, 2, 3],

'postTestScore': [25, 94, 57, 62, 70]
})
```

```python
df.columns
```

```python
plt.scatter(df['preTestScore'],df['postTestScore'],s=df.age)
plt.show()
```

```python
plt.scatter(df['preTestScore'],df['postTestScore'],s=300,c = df.female)
plt.show()
```
