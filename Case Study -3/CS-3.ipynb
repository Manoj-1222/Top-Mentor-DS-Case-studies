```python
def match_words_count(words):
    count = 0
    for word in words:
        word = word.lower()
        if(len(word)>1 and word[0]==word[-1]):
            count +=1
    return count
words = ["Koooook","Kim jon","John raj"]
print(match_words_count(words))
```

```python
def last(n): return n[-1]
def tuple_sort(tuples):
    return sorted(tuples,key=last)
print(tuple_sort([(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]))
```

```python
def long_words(n,strings):
    return [word for word in strings.split(" ") if len(word)>n]
print(long_words(3, "The quick brown fox jumps over the lazy dog"))
```

```python
def common_data(list1,list2):
    return not len(set(list1).intersection(list2))==0
print(common_data([1,2,3,4,5], [5,6,7,8,9]))

print(common_data([1,2,3,4,5], [6,7,8,9]))
```

```python
# Enumerate() method adds a counter to an iterable and returns it in a form of enumerate object. This enumerate object can then be used directly in for loops or be converted into a list of tuples using list() method.
color = ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']
color = [x for (index,x) in enumerate(color) if index not in (0,4,5)]
print(color)
```
