```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('middle_tn_schools.csv')
print(data.columns)
display(data.describe())
```

```python
display(data.info())
```

```python
display(data[['reduced_lunch','school_rating']].groupby(['school_rating']).describe())
```

```python
display(data[['reduced_lunch','school_rating']].corr())
```

```python
plt.scatter(data['reduced_lunch'],data['school_rating'],s=14,c='r')
plt.xlabel('Reduced Lunch Schools')
plt.ylabel('School Rating')
plt.show()
```

```python
import seaborn as sns
sns.regplot(data,x='reduced_lunch',y='school_rating')
plt.show()
```

```python
df1 = data.drop(['name','school_type'],axis=1)
corr = df1.corr()
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns)
plt.show()
```
