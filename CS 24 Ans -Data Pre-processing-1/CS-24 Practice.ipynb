```python
prisoners.loc['Total'] = prisoners.sum(axis=0)
```

```python
plt.rcParams["figure.figsize"]=(15,7)
```

```python
states = prisoners.iloc[:-1,0].values
totals = prisoners.iloc[:-1,-1].values
plt.xticks(np.arange(len(states)),labels=states,rotation=90)
plt.xticks
plt.bar(states,totals)
plt.show()
```

```python
xlabels = prisoners['STATE/UT'].values
plt.figure(figsize=(20, 10))

plt.xticks(np.arange(xlabels.shape[0]),xlabels,rotation='vertical', fontsize=18)
plt.xticks

plt.bar(np.arange(prisoners.iloc[:-1].values.shape[0]),prisoners['Total Benifitted'].iloc[:-1],align='edge')
plt.show()
```
