```python
def histogram(items):
    for n in items:
        while (n!=0):
            print("@",end='')
            n-=1
        print()
histogram([1,2,3])
```

```python
def list_concatenate(list):
    result =''
    for item in list:
        result += str(item)
    return result
print(list_concatenate([1,'a',2,3,4]))
```

```python
# Computing Eucledian Distance
import math
def dist(p1,p2):
    return math.sqrt(((p1[0]-p2[0])**2) + ((p1[1]-p2[1])**2))
p1 = [0, 0]
p2 = [6, 6]
print(dist(p1,p2))
```
