Business Requirement 2

Which Job Title in Year 2014 has highest mean salary?

```python
mean_salary_title_2014 = salary.groupby(['Year','JobTitle'])["TotalPayBenefits"].mean()
```

```python
mean_salary_title_2014
```

```python
z = salary.query('Year == 2014')
z = z.filter(['JobTitle','TotalPayBenefits'])
means = z.groupby('JobTitle').mean()
highest_salary_mean = means[means["TotalPayBenefits"]==means['TotalPayBenefits'].max()]
print(highest_salary_mean)
# means.sort_values(by='TotalPayBenefits',ascending=False).iloc[:,-1]
```
