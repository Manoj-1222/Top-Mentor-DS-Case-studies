```python
# Write a NumPy program to create a 3x3 matrix with values ranging from 2 to 10.  
import numpy as np
matrix = np.arange(2,11).reshape(3,3)
print(matrix)
```

```python
# Write a NumPy program to convert an array to a float type.  
a=[1,2,3,4,4]
print("Original Array:",a)
float_array = np.array(a,dtype=float)
print(float_array)
```

```python
one_border = np.ones((5,5))
one_border[1:-1,1:-1] = 0
print(one_border)
```

```python
z = np.ones((4,4))
z = np.pad(z,(1,1),mode='constant',constant_values=0)
print(z)
```

```python
chess = np.zeros((8,8))
chess[1::2,::2]=1
chess[::2,1::2]=1
print(chess)
```

```python
print(np.empty((3,3)))
print(np.full((3,2),2))
```

```python
fvalues = [0, 12, 45.21, 34, 99.91]
F = np.array(fvalues)
centigrade = 5*F/9-5*32/9
print(centigrade)
```
